## Section 1 


 #### Import Libraries
 

```python
import pandas as pd
import numpy as np
import requests
```

```python

toronto_wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
read_toronto_wiki = requests.get(toronto_wiki)

nghbrhd_data = pd.read_html(read_toronto_wiki.content, header = 0)[0]
nghbrhd_data

# Display the raw data
```

Displaying the data with Neighborhood excluding "Not Assigned"

```python
toronto_df = nghbrhd_data[nghbrhd_data.Neighbourhood != 'Not assigned']
toronto_df.reset_index(inplace = True)
toronto_df.head()
```

```python

print('Shape of the Dataframe :', toronto_df.shape)
print('Unique Neighborhoods :' ,len(toronto_df['Postal Code'].unique()))
toronto_df.groupby(['Postal Code']).first()
 
```

```python
url = 'http://cocl.us/Geospatial_data'

df_geocodes=pd.read_csv(url)
df_geocodes.head()
```

```python
print(df_geocodes.dtypes)
print(df_geocodes.shape)
df_toronto_postal = toronto_df.join(df_geocodes.set_index('Postal Code'), on='Postal Code')
df_toronto_postal
                                    

                                    
```

```python

df_toronto_postal = df_toronto_postal.reset_index()
df_toronto_postal.drop(['index'], axis = 'columns', inplace = True)
df_toronto_postal = df_toronto_postal.set_index('level_0')
toronto_postal_df = df_toronto_postal.rename(index = {'level_0' : 'index'})
toronto_postal_df.index.name = 'index'
print(toronto_postal_df.shape)
toronto_postal_df
```
